```
from pyspark.sql import SparkSession
```

```
spark = SparkSession.builder.appName('aggs').getOrCreate()
```

```
df = spark.read.csv('/FileStore/tables/sales_info.csv', inferSchema=True, header=True)
```

```
df.show()
```

```
df.printSchema()
```

GroupBy funtion
---

```
df.groupBy("Company")
```

```
# 회사(Company)별 평균(mean) 매출액(Sales) 보기
df.groupBy("Company").mean().show()
```

```
# 회사(Company)별 총(sum) 매출액(Sales) 보기
df.groupBy("Company").sum().show()
```

```
# 회사(Company)별 최대(max) 매출액(Sales) 보기
df.groupBy("Company").max().show()
```

```
# 회사(Company)별 최소(min) 매출액(Sales) 보기
df.groupBy("Company").min().show()
```

```
# 회사(Company)별 직원(people) 수(count) 보기
df.groupBy("Company").count().show()
```

Aggregate function
---
딕셔너리 형태로 key 값엔 연산을 적용할 컬럼값을, value 값엔 컬럼에 적용할 연산 입력
=> df.agg({'Columns':'function'})

```
# Sales열의 합계(sum)
df.agg({'Sales':'sum'}).show()
```

```
# Sales열의 최댓값(max)
df.agg({'Sales':'max'}).show()
```

```
group_data = df.groupBy("Company")
```

```
# df.groupBy('Company)'.max().show()와 동일한 결과 출력
group_data.agg({'Sales':'max'}).show()
```

```
from pyspark.sql.functions import countDistinct, avg, stddev
```

```
df.select(countDistinct('Sales')).show()
```

```
df.select(avg('Sales')).show()
```

```
df.select(avg('Sales').alias('Average_Sales')).show()
```

```
df.select(stddev('Sales')).show()
```

```
from pyspark.sql.functions import format_number
```

```
sales_std = df.select(stddev("Sales").alias('std'))
```

```
sales_std.show()
```

```
sales_std.select(format_number('std', 2)).show()
```

```
sales_std.select(format_number('std', 2).alias('std')).show()
```
