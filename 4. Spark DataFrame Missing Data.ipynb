```
from pyspark.sql import SparkSession
```

```
spark = SparkSession.builder.appName('miss').getOrCreate()
```

```
df = spark.read.csv('/FileStore/tables/ContainsNull.csv', header=True, inferSchema=True)
```

```
df.show()
```

```
# null값이 하나라도 포함된 행(row) 제거
df.na.drop().show()
# df.na.drop(how='any').show()
```

```
# null값 drop시에 임계값 설정 
df.na.drop(thresh=2).show()
```

```
# 모든 값이 null값인 행(row) 제거
df.na.drop(how='all').show()
```

```
# 특정 범주의 null값만 제거
df.na.drop(subset=['Sales']).show()
```

```
df.printSchema()
```

spark에서 null값을 대체할때 자동으로 같은 타입만 변환해줌!

```
# null값을 'Fill Value'로 대체 -> string인 null 값만 변환
df.na.fill('Fill Value').show()
```

```
# null값을 '0'로 대체 -> 숫자인 null 값만 변환
df.na.fill(0).show()
```

```
# 특정 범주의 null값만 지정해서 변환
df.na.fill('No Name', subset=['Name']).show()
```

```
from pyspark.sql.functions import mean
```

```
mean_val = df.select(mean(df['Sales'])).collect()
```

```
mean_sales = mean_val[0][0]
```

```
df.na.fill(mean_sales, ['Sales']).show()
```
